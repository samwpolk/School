```python
#classic graident descent
import numpy as np
import matplotlib.pyplot as plt
import random

epsilon = .000001
beta = .9
lr = .5
VdWx = 1
VdWx = 1
max_it = 10000
lossind = []
lossdepx = []
lossdepy = []
dfindx = []
dfdepy = []
priorx = []
priory = []


def f(x,y):
   
   return (25. - np.square(x - 3.) - np.square(y - 2.))
    
    
```

```python
def df(x,y):
    
    return   [(2*(x - 3))/f(x,y),(2*(y - 2))/f(x,y)]

    
```

```python
xlist = np.linspace(0., 6., 60.)
ylist = np.linspace(0.67, 3.5, 60.)
X, Y = np.meshgrid(xlist, ylist)

fig, ax = plt.subplots()
Z =  -(25. - np.square(X - 3.) - np.square(Y - 2.))
cp = ax.contourf(X, Y, Z)
ax.clabel(cp, inline=True, 
          fontsize=7)
ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(cp)
plt.show()
```

```python
#x = random.random()
#y = random.random()
#x,y = [0.7014897258404872, 0.23873965536031538]
x,y = [0., 0.]

print("starting values")
print([x,y])
print()


for c in range(max_it):
    '''if ( c < 1000):
        gamma = .1
    else:
        gamma = .01'''
        
    dfx,dfy = df(x,y)
    tx = x
    ty = y
   # VdWx = beta * VdW + (1 - beta ) * dfx
   # VdWy = beta * VdW + (1 - beta ) * dfy
    x = x - lr *  dfx
    y = y - lr *  dfy
    dfindx.append(dfx)
    dfdepy.append(dfy)
    priorx.append(x)
    priory.append(y)
    lossind.append(c)
    lossdepx.append(x - tx)
    lossdepy.append(y - ty)
   
    #print([np.abs(x - tx) ,np.abs(y - ty), f(x,y),c])
    if (np.abs(x - tx) < epsilon and np.abs(y - ty) < epsilon):
        break
   
   
```

```python
results = [-np.sqrt(f(x,y)),x,y,c]
print("extrema ",results[0])
print('x is ',results[1])
print('y is ',results[2])
print('epochs ',results[3])
```
