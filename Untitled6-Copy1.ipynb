```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/Users/swp/Downloads/02Rice.png')
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F, ksize = 11)
fig, aux = plt.subplots(figsize = (5,5))
aux.imshow(laplacian, cmap='gray')
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F, ksize = 23)
fig, aux = plt.subplots(figsize = (4,4))
aux.imshow(laplacian, cmap='gray')
```

```python
canny = cv2.Canny(img,25,150)
fig, aux = plt.subplots(figsize = (4,4))
aux.imshow(canny, cmap='gray')
```

```python
fig, aux = plt.subplots(figsize = (5,5))
image_gr = im.convert("L")
edges = cv2.Canny(img,100,200)
arr = np.asarray(edges)
aux.imshow(edges,cmap='gray')
grad = signal.convolve2d(arr,k2,mode='same',boundary='symm')
```
