```python
#Dynamic Gamma
import numpy as np
import matplotlib.pyplot as plt
import random


epsilon = 0.0000001
max_it = 100000
lossind = []
lossdepx = []
lossdepy = []
dfindx = []
dfdepy = []

def f(x,y):
   
   return (25. - np.square(x - 3.) - np.square(y - 2.))
    
    
```

```python
def df(x,y):
    
    return [(2*(x - 3))/f(x,y),(2*(y - 2))/f(x,y)]


    
    
```

```python
#x = random.random()
#y = random.random()
x,y = [0.7014897258404872, 0.23873965536031538]
print("starting values")
print([x,y])
print()


for c in range(max_it):
    if ( c < max_it/2):
        gamma = .5
    else:
        gamma = gamma/10
        
    dfx,dfy = df(x,y)
    tx = x
    ty = y
    y = y - gamma * dfy
    x = x - gamma * dfx
    
    dfindx.append(dfx)
    dfdepy.append(dfy)
    lossind.append(c)
    lossdepx.append(x - tx)
    lossdepy.append(y - ty)
   
    #print([np.abs(x - tx) ,np.abs(y - ty), f(x,y),c])
    if (np.abs(x - tx) < epsilon and np.abs(y - ty) < epsilon):
        break
   
   
```

```python
plt.plot(lossind,dfindx)
```

```python
plt.plot(lossind,dfdepy)
```

```python
results = [-np.sqrt(f(x,y)),x,y,c]
print("extrema ",results[0])
print('x is ',results[1])
print('y is ',results[2])
print('epochs ',results[3])
```

```python
plt.figure(1, figsize=(7, 5))
plt.plot(lossind,lossdepx, label = "x")
plt.plot(lossind,lossdepy,label = "y")
plt.legend()
plt.xlabel('Number of Epochs')
plt.ylabel('loss')
plt.title('Dyanamic Gamma Gradient Descent')
plt.grid(True)
plt.style.use('ggplot')
plt.show()


```
