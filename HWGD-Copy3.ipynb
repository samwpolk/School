```python
#Momentum
import numpy as np
import matplotlib.pyplot as plt
import random

gamma = .18
lr = .5
Vtx = .012
Vty = .077029
epsilon = 0.000001
max_it = 100000
dfindx = []
dfdepy = []
lossind = []
lossdepx = []
lossdepy = []
priorx = []
priory = []


def f(x,y):
   
   return (25. - np.square(x - 3.) - np.square(y - 2.))
    
    
```

```python
def df(x,y):
    
    return [(2*(x - 3))/f(x,y),(2*(y - 2))/f(x,y)]

    
```

```python
#x = random.random()
#y = random.random()
x,y = [0.7014897258404872, 0.23873965536031538]
print("starting values")
print([x,y])
print()


for c in range(max_it):
    '''if ( c < 1000):
        gamma = .1
    else:
        gamma = .01'''
        
    dfx,dfy = df(x,y)
    tx = x
    ty = y
    Vtx = gamma * Vtx + lr * dfx
    Vty = gamma * Vtx + lr * dfy
    x = x - Vtx
    y = y - Vty
    priorx.append(x)
    priory.append(y)
    dfindx.append(dfx)
    dfdepy.append(dfy)
    lossind.append(c)
    lossdepx.append(x - tx)
    lossdepy.append(y - ty)
   
    #print([np.abs(x - tx) ,np.abs(y - ty), f(x,y),c])
    if (np.abs(x - tx) < epsilon and np.abs(y - ty) < epsilon):
        break
   
   
```

```python
results = [-np.sqrt(f(x,y)),x,y,c]
print("extrema ",results[0])
print('x is ',results[1])
print('y is ',results[2])
print('epochs ',results[3])
```

```python
plt.figure(1, figsize=(7, 5))
plt.plot(lossind,lossdepx, label = "x")
plt.plot(lossind,lossdepy,label = "y")
plt.legend()
plt.xlabel('Number of Epochs')
plt.ylabel('loss')
plt.title('Naive Gradient Descent')
plt.grid(True)
plt.show()


```
